## Exercise - Create a Calculator
Create a calculator capable of performing addition, subtraction, multiplication and division operations on two numbers. Your program should format the output in a readable manner!

```python
a=int(input("Enter a Number: "))
b=int(input("Enter a Second Number: "))
c=a+b
d=a-b
e=a*b
f=a**b
g=a/b
h=a//b
i=a%b
print("Addition of two no. is: ",c)
print("Subtraction of two no. is: ",d)
print("Multiplication of two no. is: ",e)
print("Exponential of two no. is: ",f)
print("Division of two no. is: ",g)
print("Floor Division of two no. is: ",h)
print("Modulus of two no. is: ",i)
```
